```python
import http.client
import json
import requests
import pandas as pd
from sys import exit
from json import loads
```

```python
conn = http.client.HTTPSConnection('dominio')
```

```python
headers = {
    'Authorization': "***",
    'App-Token': "E***"
}
```

```python
conn.request("GET", "/apirest.php/initSession", "", headers)
init_res = conn.getresponse().read().decode("utf-8")
init_body = loads(init_res)
session_token = init_body["session_token"]
print(session_token)
```

```python
headers["Session-Token"] = session_token
```

```python
conn.request("GET", "/apirest.php/Ticket?range=0-1000000", "", headers)
#conn.request("GET", "/apirest.php/User?range=0-1000000", "", headers)

res = conn.getresponse().read().decode("utf-8")
tickets = loads(res)
print(tickets)
```

```python
df = pd.DataFrame(tickets)
```

{x['id']:x['name'] for x in tickets}




```python
df = df.to_csv('ticketfinais.csv', sep=';')
```
